```python
df3['a'].hist( grid=False, color='blue', edgecolor='black', linewidth=1.2)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# color='red', edgecolor='white', linewidth=1.2, bins=25, alpha=0.4 
df3['a'].plot.hist(color='red', edgecolor='white', linewidth=1.2, bins=30, alpha=0.4 )
```

```python
#Create a boxplot comparing the a and b columns.
df3[['a', 'b']].plot.box()
```

```python
#Create a kde plot of the 'd' column
df3['d'].plot.kde(color='red')
```

```python
#Figure out how to increase the linewidth and make the linestyle dashed.
#(Note: You would usually not dash a kde plot line)
df3['d'].plot.kde(color='red', linewidth=5, linestyle='--')
```

```python
#Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).
df3.iloc[0:30].plot.area(alpha=0.4)
```
